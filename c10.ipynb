```python
import numpy as np
import matplotlib.pyplot as plt
from c10.c10_util import *
from c10.Height import Height
```

```python
xs_male, xs_female = load_height('./c10/variability_data_1000.pkl')
xs_male = jitter(xs_male)
xs_female = jitter(xs_female)
```

```python
mus, sigmas = find_prior_ranges(xs_male)

suite = Height(mus, sigmas, name='male')
suite.update_set(xs_male)
print(suite.max_likelihood())

cdf_male = suite.make_cv_cdf()
```

```python
mus, sigmas = find_prior_ranges(xs_female)

suite = Height(mus, sigmas, name='female')
suite.update_set(xs_female)
print(suite.max_likelihood())

cdf_female = suite.make_cv_cdf()
```

```python
mus, sigmas = find_prior_ranges(xs_male)

suite = Height(mus, sigmas)
suite.log()
suite.log_update_set(xs_male)
suite.exp()
suite.normalize()
print(suite.max_likelihood())
```

```python
mus, sigmas = find_prior_ranges(xs_male)

suite = Height(mus, sigmas)
suite.log()
suite.log_update_set_mean_var(xs_male)
suite.exp()
suite.normalize()
print(suite.max_likelihood())
```

```python
mus, sigmas = find_prior_ranges(xs_male, median_flag=True)

suite = Height(mus, sigmas, name='male')
suite.log()
suite.log_update_set_median_ipr(xs_male)
suite.exp()
suite.normalize()
print(suite.max_likelihood())
cdf_male = suite.make_cv_cdf()
suite.plot()
```

```python
mus, sigmas = find_prior_ranges(xs_female, median_flag=True)

suite = Height(mus, sigmas, name='female')
suite.log()
suite.log_update_set_median_ipr(xs_female)
suite.exp()
suite.normalize()
print(suite.max_likelihood())
cdf_female = suite.make_cv_cdf()
suite.plot()
```

```python
cdf_male.plot_with([cdf_female])
```
